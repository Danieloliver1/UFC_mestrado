```python
# Esta função cria os grafos no Neo4j todo o projeto.
def create_graph_from_dataframe(df, driver):
    """
    Creates Neo4j graphs for general, viral, and misinformation messages from a DataFrame.

    Parameters:
        df (pd.DataFrame): The DataFrame containing WhatsApp message data.
        driver (neo4j.Driver): The Neo4j driver for database connection.

    Returns:
        dict: A dictionary with the number of nodes and edges for each graph type.
    """
    # Preprocess DataFrames
    df_general = df[df["messenger"] == "whatsapp"].groupby(
        ["id_member_anonymous", "id_group_anonymous"]
    ).agg(messages=("score_sentiment", "size")).reset_index()

    df_viral = df[(df["messenger"] == "whatsapp") & (df["score_sentiment"] > 0.5)].groupby(
        ["id_member_anonymous", "id_group_anonymous"]
    ).agg(messages=("score_sentiment", "size")).reset_index()

    df_misinformation = df[(df["messenger"] == "whatsapp") & (df["score_misinformation"] > 0.5)].groupby(
        ["id_member_anonymous", "id_group_anonymous"]
    ).agg(messages=("score_misinformation", "size")).reset_index()

    # Function to create nodes and relationships in Neo4j
    def create_nodes_and_relationships(df_subset, relationship_type):
        with driver.session() as session:
            # Create users and groups as nodes
            unique_users = df_subset["id_member_anonymous"].unique()
            unique_groups = df_subset["id_group_anonymous"].unique()

            # Create nodes for users
            for user in unique_users:
                session.run("MERGE (u:User {id: $id})", {"id": user})

            # Create nodes for groups
            for group in unique_groups:
                session.run("MERGE (g:Group {id: $id})", {"id": group})

            # Create relationships
            for _, row in df_subset.iterrows():
                session.run("""
                    MATCH (u:User {id: $id_member_anonymous})
                    MATCH (g:Group {id: $id_group_anonymous})
                    MERGE (u)-[r:MESSAGED {type: $type}]->(g)
                    ON CREATE SET r.weight = $messages
                    ON MATCH SET r.weight = r.weight + $messages
                """, {
                    "id_member_anonymous": row["id_member_anonymous"],
                    "id_group_anonymous": row["id_group_anonymous"],
                    "type": relationship_type,
                    "messages": row["messages"]
                })

    # Create graphs for each message type
    create_nodes_and_relationships(df_general, "general")
    create_nodes_and_relationships(df_viral, "viral")
    create_nodes_and_relationships(df_misinformation, "misinformation")

    # Analysis: Count nodes and edges for each graph
    stats = {}
    with driver.session() as session:
        for rel_type in ["general", "viral", "misinformation"]:
            node_count = session.run("""
                MATCH (n) RETURN COUNT(DISTINCT n) AS num_nodes
            """).single()["num_nodes"]

            edge_count = session.run(f"""
                MATCH ()-[r:MESSAGED {{type: '{rel_type}'}}]->() RETURN COUNT(r) AS num_edges
            """).single()["num_edges"]

            stats[rel_type] = {"nodes": node_count, "edges": edge_count}

    return stats

```

```python


def get_graph_metrics(driver):
    with driver.session() as session:
        # Quantidade de nós (usuários)
        num_users = session.run("MATCH (u:User) RETURN COUNT(u) AS num_users").single()["num_users"]

        # Grau de centralidade e força para cada tipo de mensagem
        metrics = {"Users": num_users}

        # Graus e forças gerais número total de relacionamentos
        centrality_query = """
            MATCH (u:User)-[r:MESSAGED]->()
            WHERE r.type = 'general'
            RETURN u.id AS user_id, COUNT(r) AS degree_general, SUM(r.weight) AS strength_general
        """
        general_metrics = session.run(centrality_query)

        for record in general_metrics:
            metrics[f"{record['user_id']}_degree_general"] = record["degree_general"]
            metrics[f"{record['user_id']}_strength_general"] = record["strength_general"]

        # Graus e forças virais número total de relacionamentos
        centrality_query_viral = """
            MATCH (u:User)-[r:MESSAGED]->()
            WHERE r.type = 'viral'
            RETURN u.id AS user_id, COUNT(r) AS degree_viral, SUM(r.weight) AS strength_viral
        """
        viral_metrics = session.run(centrality_query_viral)

        for record in viral_metrics:
            metrics[f"{record['user_id']}_degree_viral"] = record["degree_viral"]
            metrics[f"{record['user_id']}_strength_viral"] = record["strength_viral"]

        # Graus e forças de desinformação número total de relacionamentos
        centrality_query_misinformation = """
            MATCH (u:User)-[r:MESSAGED]->()
            WHERE r.type = 'misinformation'
            RETURN u.id AS user_id, COUNT(r) AS degree_misinformation, SUM(r.weight) AS strength_misinformation
        """
        misinformation_metrics = session.run(centrality_query_misinformation)

        for record in misinformation_metrics:
            metrics[f"{record['user_id']}_degree_misinformation"] = record["degree_misinformation"]
            metrics[f"{record['user_id']}_strength_misinformation"] = record["strength_misinformation"]

        return metrics

# Chamando a função
data = get_graph_metrics(driver)
```

```python
import networkx as nx
import matplotlib.pyplot as plt

def obter_dados_com_rotulos(session):
    consulta = "MATCH (n)-[r]->(m) RETURN n.id AS source, r.type AS relation, m.id AS target  LIMIT 1000"
    resultado = session.run(consulta)
    grafo = nx.DiGraph()
    rotulos = {}  # Mapeamento de IDs para rótulos amigáveis
    contador = 1

    for registro in resultado:
        source = registro["source"]
        target = registro["target"]
        relation = registro["relation"]

        # Criar rótulos curtos para cada nó
        if source not in rotulos:
            rotulos[source] = f"N{contador}"
            contador += 1
        if target not in rotulos:
            rotulos[target] = f"N{contador}"
            contador += 1

        # Adicionar aresta ao grafo com rótulos curtos
        grafo.add_edge(rotulos[source], rotulos[target], relation=relation)

    return grafo, rotulos
```

```python
# exemplo de grafos em gráficos para exibir

with driver.session() as session:
    G, rotulos = obter_dados_com_rotulos(session)

# Desenhar o grafo
plt.figure(figsize=(16, 12))
pos = nx.spring_layout(G)  # Layout ajustado para melhor visualização
nx.draw(
    G,
    pos,
    with_labels=True,
    labels={node: node for node in G.nodes()},  # Usar os rótulos curtos
    node_color="lightblue",
    edge_color="gray",
    node_size=500,
    font_size=10,
)
nx.draw_networkx_edge_labels(
    G,
    pos,
    edge_labels={(u, v): d['relation'] for u, v, d in G.edges(data=True)},
    font_color="red",
    font_size=8,
)

plt.show()

# Exibir mapeamento de IDs originais para rótulos curtos
# print("Mapeamento de IDs para rótulos curtos:")
# for id_original, rotulo in rotulos.items():
#     print(f"{rotulo}: {id_original}")
```
