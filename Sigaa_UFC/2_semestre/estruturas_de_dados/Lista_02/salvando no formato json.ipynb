```python
import json

def save_user_graph_to_json(df, rede_social, json_file_path):
    """
    Salva grafos direcionados entre usuários no formato JSON para mensagens gerais, virais e de desinformação.

    Parâmetros:
        df (pd.DataFrame): DataFrame contendo dados de mensagens do WhatsApp.
        rede_social (str): Rede social a ser filtrada (ex.: 'WhatsApp').
        json_file_path (str): Caminho do arquivo JSON onde os dados serão salvos.

    Retorna:
        None
    """
    # Validar se a rede social existe no DataFrame
    if rede_social not in df["messenger"].unique():
        raise ValueError(f"Rede social '{rede_social}' não encontrada no DataFrame.")

    # Filtrar dados para a rede social especificada
    df_filtered = df[df["messenger"] == rede_social]

    # Função para criar nós e arestas direcionadas
    def create_edges_json(df_subset, relationship_type):
        """
        Cria nós e arestas direcionadas no formato JSON.

        Parâmetros:
            df_subset (pd.DataFrame): Subconjunto de dados para criar nós e arestas.
            relationship_type (str): Tipo de relacionamento ('general', 'viral', 'misinformation').

        Retorna:
            dict: Dicionário contendo nós e arestas.
        """
        nodes = {}
        edges = []

        for group_id in df_subset["id_group_anonymous"].unique():
            group_members = df_subset[df_subset["id_group_anonymous"] == group_id]
            senders = group_members.groupby("id_member_anonymous").agg(
                total_messages=("messages", "sum")
            ).reset_index()

            # Adicionar o grupo como nó
            nodes[group_id] = {"id": group_id, "type": "Group"}

            # Adicionar nós e arestas ao dicionário
            for _, sender_row in senders.iterrows():
                sender = sender_row["id_member_anonymous"]
                total_messages = sender_row["total_messages"]

                # Adicionar o remetente como nó
                nodes[sender] = {"id": sender, "type": "User"}

                for recipient in group_members["id_member_anonymous"].unique():
                    if sender != recipient:
                        # Adicionar o destinatário como nó
                        nodes[recipient] = {"id": recipient, "type": "User"}

                        # Criar a aresta entre remetente e destinatário
                        edges.append({
                            "source": sender,
                            "target": recipient,
                            "relationship_type": relationship_type,
                            "weight": total_messages
                        })

        return {"nodes": list(nodes.values()), "edges": edges}

    # Criar grafos para mensagens gerais, virais e de desinformação
    df_general = df_filtered.groupby(
        ["id_member_anonymous", "id_group_anonymous"]
    ).agg(messages=("score_sentiment", "size")).reset_index()

    df_viral = df_filtered[df_filtered["score_sentiment"] > 0.5].groupby(
        ["id_member_anonymous", "id_group_anonymous"]
    ).agg(messages=("score_sentiment", "size")).reset_index()

    df_misinformation = df_filtered[df_filtered["score_misinformation"] > 0.5].groupby(
        ["id_member_anonymous", "id_group_anonymous"]
    ).agg(messages=("score_misinformation", "size")).reset_index()

    # Combinar os grafos em um único dicionário
    graph_data = {
        "general": create_edges_json(df_general, "general"),
        "viral": create_edges_json(df_viral, "viral"),
        "misinformation": create_edges_json(df_misinformation, "misinformation")
    }

    # Salvar os dados no arquivo JSON
    with open(json_file_path, 'w') as f:
        json.dump(graph_data, f, indent=4)

# Exemplo de uso:
# save_user_graph_to_json(df, "WhatsApp", "user_graph.json")
```
